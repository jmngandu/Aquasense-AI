```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import random

# Load the water quality dataset
url = 'https://raw.githubusercontent.com/jmngandu/Aquasense-AI/main/models/water_potability.csv'
df = pd.read_csv(url)
print(df.head())
print("\nDataset information:")
print(df.info())
print("\nMissing values in the dataset:")
print(df.isnull().sum())
df = df.fillna(df.mean())

# Define normal ranges for each parameter (example ranges)
normal_ranges = {
    'ph': (6.5, 8.5),
    'Hardness': (150, 300),
    'Solids': (300, 10000),
    'Chloramines': (0, 4),
    'Sulfate': (25, 250),
    'Conductivity': (50, 1500),
    'Organic_carbon': (0, 20),
    'Trihalomethanes': (0, 80),
    'Turbidity': (0, 5)
}

# Label the data
def label_row(row):
    for param, (low, high) in normal_ranges.items():
        if row[param] < low or row[param] > high:
            return 1  # Out of normal range
    return 0  # Normal range

df['Label'] = df.apply(label_row, axis=1)

# Split the data
X = df.drop(['Label'], axis=1)
y = df['Label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Evaluate the model
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

df['Prediction'] = clf.predict(X)

pop_url = 'https://raw.githubusercontent.com/jmngandu/Aquasense-AI/main/models/Population-of-Cross-River-State-2022-Projection.csv'
pop_df = pd.read_csv(pop_url)
pop_df['Population'] = pop_df['Population'].str.replace('[^\d]', '', regex=True).astype(int)

locations = pop_df['City'].tolist()
df['Inspection_Location'] = df.apply(
    lambda x: random.choice(locations) if x['Prediction'] == 1 else np.nan, axis=1
)

# Visualization of the population by city in Cross River State
plt.figure(figsize=(10, 6))
plt.bar(pop_df['City'], pop_df['Population'], color='darkblue')
plt.xlabel('Cities')
plt.ylabel('Population')
plt.title('Population of Different Cities in Cross River State')
plt.xticks(rotation=45)
plt.show()
```

```python
# Function to simulate sending a ping
def send_ping(row):
    if row['Prediction'] == 1:
        print(f"Ping sent! Inspection needed at {row['Inspection_Location']} for out-of-range parameters.")
    return row

# Apply the function to rows requiring inspection
df = df.apply(send_ping, axis=1)

print("\nRows with out-of-range values requiring inspection:")
print(df[df['Prediction'] == 1])
```
